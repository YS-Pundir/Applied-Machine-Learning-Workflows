```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
from IPython.display import display
```

```python
# 1. Car Dataset (Predicting car price)
data = {
    "Brand": ["Maruti", "Hyundai", "Maruti", "Honda", "Toyota", "Honda"],
    "Year": [2018, 2017, 2018, 2016, 2019, 2015],
    "Fuel": ["Petrol", "Diesel", "Petrol", "Petrol", "Diesel", "Petrol"],
    "KMs_Driven": [30000, 40000, 30000, None, 20000, 70000],
    "Price": [400000, 500000, 400000, 450000, 800000, -200000], # Notice: negative price, duplicate Maruti
}
df_car = pd.DataFrame(data)

# 2. Student Dataset (Understanding student performance)
data_student = {
    "Student_ID": ["S01", "S02", "S01", "S03", "S04"],
    "Marks": [85, 72, 85, None, 90],
    "Attendance": [90, 88, 90, 95, 92],
    "Grade": ["A", "B", "A", "C", -10], # Notice: weird -10 Grade, duplicate S01
}
df_student = pd.DataFrame(data_student)

# 3. E-commerce Dataset (Analyzing orders and pricing)
data_ecommerce = {
    "OrderID": ["O100", "O101", "O100", "O102", "O103"],
    "Product": ["Phone", "Laptop", "Phone", "Phone", "Laptop"],
    "Quantity": [1, 2, 1, None, 1],
    "Price": [20000, 60000, 20000, 15000, -5000], # Notice: negative price, missing quantity
}
df_ecom = pd.DataFrame(data_ecommerce)

print("Datasets loaded!")
```

Inspecting the dataset

```python
print("car Dataset")
print(df_car.head(5))
print("Structral summary of car data")
print(df_car.info())
print("Statictical summary of car data ")
print(df_car.describe())
print()
```

```python
#Cleaning the data
df_car=df_car.fillna(df_car["KMs_Driven"].mean())
df_car["Price"]=df_car["Price"].abs()
df_car=df_car.drop_duplicates()
print("Updated car dataset")
display(df_car)
print("<"*60)
# encoding
label_encoder = LabelEncoder()
df_car['Fuel'] = label_encoder.fit_transform(df_car['Fuel'])
print("Petrol=1,diesel=0")
df_car_encoded=pd.get_dummies(df_car,dtype=int,columns=["Brand"])
df_car_encoded
print("Encoded Dataset")
display(df_car_encoded)
# Scaling
columns_to_scale=["Year","KMs_Driven","Price"]
scaler=StandardScaler()
df_car_scaled=df_car_encoded.copy()
df_car_scaled[columns_to_scale]=scaler.fit_transform(df_car_encoded[columns_to_scale])
df_car_scaled=pd.DataFrame(df_car_scaled)
print("Scaled Dataset")
display(df_car_scaled)
# Train-Test spliting
x=df_car_scaled.drop("Price",axis=1)
y=df_car_scaled["Price"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4, random_state=42)
```

Inspecting the student dataset

```python
print("-"*60)
print("Student Dataset")
display(df_student)
print("Structral summary of student data")
display(df_student.info())
print("Statictical summary of student data ")
print(df_student.describe())
print()
```

```python
# Correcting the false data 
df_student["Grade"]=df_student["Grade"].replace({-10:np.nan,"-10":np.nan})
df_student["Marks"]=df_student["Marks"].abs()
df_student["Attendance"]=df_student["Attendance"].abs()
#handelling miissing values and duplicates
df_student["Marks"]=df_student["Marks"].fillna(df_student["Marks"].mean())
df_student["Grade"]=df_student["Grade"].fillna(df_student["Grade"].mode()[0])
df_student=df_student.drop_duplicates()
#Droping the id column 
df_student=df_student.drop("Student_ID",axis=1)
# encoding the columns
df_student_encoded=df_student.copy()
df_student_encoded=pd.get_dummies(df_student,dtype=int,columns=["Grade"])
#scaling
columns_to_scale=["Marks","Attendance"]
scaler=StandardScaler()
df_student_scaled=df_student_encoded.copy()
df_student_scaled[columns_to_scale]=scaler.fit_transform(df_student_encoded[columns_to_scale])
df_student_scaled=pd.DataFrame(df_student_scaled)
#Train,test spliting
x=df_student_scaled.drop("Marks",axis=1)
y=df_student_scaled["Marks"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```
