# Part 1 : Data Leakge 
## Case 1 : Leakage via missing value 
If we have missing values in data and we tried to fill  it on entire dataset before filling it , then we have filled the missing values  with average of test data as well . 

```python
df_missing = pd.DataFrame({
    'Student': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Score': [40, 60, np.nan, 80, np.nan, 90]  # Missing scores for C and E
})
print("Original data : ")
display(df_missing)
# first split data into testdata and train data
df_train,df_test=train_test_split(df_missing,test_size=0.2,random_state=42)
print("Training datset")
print()
display(df_train)
print("Testign data : ")
display(df_test)
```

```python
#  NOw take the average from training data and fill those values in both of the splited data
imputer=SimpleImputer(strategy="mean")
imputer.fit(df_train[["Score"]])
df_train["Score"]=imputer.transform(df_train[["Score"]])
df_test["Score"]=imputer.transform(df_test[["Score"]])
print("Filled test data : ")
display(df_test)
print("Filled trained data : ")
display(df_train)
```
