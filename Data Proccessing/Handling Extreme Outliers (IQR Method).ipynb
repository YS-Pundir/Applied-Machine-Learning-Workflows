1.  **The IQR Method (Interquartile Range):**
    *   This is a common statistical technique to identify and handle outliers.
    *   **Quartiles:** Imagine arranging all your house prices from lowest to highest. You can divide this sorted list into four equal parts (quarters).
        *   **Q1 (First Quartile):** This is the value below which 25% of the data falls.
        *   **Q3 (Third Quartile):** This is the value below which 75% of the data falls (or above which 25% falls).
    *   **IQR (Interquartile Range):** This is the range between Q1 and Q3 (`IQR = Q3 - Q1`). It represents the middle 50% of your data and is considered the 'normal' spread of values.
    *   **Outlier Detection:** The method defines boundaries for what's considered 'normal' data:
        *   **Lower Bound:** `Q1 - 1.5 * IQR`
        *   **Upper Bound:** `Q3 + 1.5 * IQR`
    *   Any data point **below the Lower Bound** or **above the Upper Bound** is considered an outlier.

2.  **Capping Outliers:**
    *   Instead of simply removing outliers (which might mean losing valuable data), 'capping' means replacing the outlier with the boundary value. For example, if a house is priced at 5000 Lakhs, but the upper bound is 58.75 Lakhs, we replace the 5000 Lakhs with 58.75 Lakhs. This way, the extreme value is contained, but we still keep the data point.

```python
import pandas as pd
import numpy as np
houses=pd.DataFrame({
    "Housing_price":[45, 50, 48, 52, 49, 51, 55, 5000]
})

#calculating the first and third quarrtile
Q1=houses["Housing_price"].quantile(0.25)
Q3=houses["Housing_price"].quantile(0.75)
# calculating the IQR(Interquartile range)
ITR=Q3-Q1
#calculatin upperlimit and lower limit
upper=Q3+1.5*ITR
lower=Q1-1.5*ITR
#Replacing  the outliers
houses["Corrected_prices"]=np.where(houses["Housing_price"]>upper,upper,houses["Housing_price"])
houses["Corrected_prices"]=np.where(houses["Corrected_prices"]<lower,lower,houses["Corrected_prices"])
#displaying the data without the outliers
houses
```
