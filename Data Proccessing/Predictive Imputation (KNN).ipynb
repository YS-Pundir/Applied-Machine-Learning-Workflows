If a CEO's salary is missing, filling it with the company "average" is wrong. KNN Imputer looks at the K most similar rows (Neighbors) and averages their values.

```python
from sklearn.impute import KNNImputer
import pandas as pd
import numpy as np
# Dummy data: Young vs Older employees
df_employees = pd.DataFrame({
    'Age': [22, 23, 24, 45, 47, 46],
    'Experience': [1, 1, 2, 20, 22, 21],
    'Salary': [30000, 32000, np.nan, 150000, 160000, np.nan]
})
df_employees
```

```python
#Training the imputing model
imputer=KNNImputer(n_neighbors=2)
data=imputer.fit_transform(df_employees)
#Filling the values
df_filled=pd.DataFrame(data,columns=df_employees.columns)
display(df_employees)
print(">"*60)
print("The filled dataset : ")
df_filled
```
