### Join : 
1. Inner : It is kinnd of intersaction in sets .
2. Outer : 
      *  Left Join : Joining the complete left sside of table wioth the corresponding right ones . Left one will be complete .
      *  Right Join : Joing the complete right sided table  with the corresponding rows of left one . Righ one will always will be completed.
      *  Full Join : Joing both tables completely , and all rows from both of the table will be completely joined . 



### Inner Join :->

```python
use DBeaver;

select * 
from Artist;



select * 
from Artist 
inner join Album 
on Artist.ArtistId =Album.ArtistId ;

select * 
from Album 
inner join Artist 
on Album.ArtistId =Artist.ArtistId ;
```
