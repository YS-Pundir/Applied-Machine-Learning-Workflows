```python
# STEP 0: IMPORTING OUR TOOLS
import pandas as pd
import numpy as np

# Machine Learning Tools
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.pipeline import make_pipeline

print("Tools imported successfully! Let's classify some data.")
```

```python
# 1. Let's simulate a real-world dataset of 400 Students
rng = np.random.default_rng(seed=7)

study_hours  = rng.uniform(1, 10, size=400)
sleep_hours  = rng.uniform(4, 9, size=400)
distractions = rng.uniform(0, 5, size=400)

# We calculate a hidden score, and say anyone above 70 PASSES (1), otherwise FAILS (0)
hidden_scores = 40 + (6.5 * study_hours) + (1.2 * sleep_hours) - (2.0 * distractions) + rng.normal(0, 7, size=400)
pass_fail_labels = (hidden_scores >= 70).astype(int) # 1 = Pass, 0 = Fail

df_students = pd.DataFrame({
    'Study_Hours': study_hours,
    'Sleep_Hours': sleep_hours,
    'Distractions': distractions,
    'Result': pass_fail_labels
})

print("--- STUDENT DATASET (First 5 Rows) ---")
print("Notice the Target ('Result') is no longer a continuous number. It is strictly 0 or 1.")
display(df_students.head())
```

```python
# 1. Separate Features and Target
X = df_students[['Study_Hours', 'Sleep_Hours', 'Distractions']]
y = df_students['Result']

# 2. Split the Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)

# 3. Train Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train)

print("Model trained successfully!")
print(f"Training Accuracy: {model.score(X_train, y_train) * 100:.1f}%")
print(f"Testing Accuracy:  {model.score(X_test, y_test) * 100:.1f}%")
```

```python
Hard_pridiction=model.predict(X_test)
prediction_probablity=model.predict_proba(X_test)

df=pd.DataFrame({
    "Actual_results":y_test.values,
    "Predictions":Hard_pridiction,
    "prediction probablity":prediction_probablity[:,1]
})
display(df)
```

```python
print("--- 1. THE RAW CONFUSION MATRIX ---")
cm = confusion_matrix(y_test, Hard_pridiction)
print(cm)

# Let's unpack these numbers into variables so we can understand them
tn = cm[0][0]  # True Negatives
fp = cm[0][1]  # False Positives
fn = cm[1][0]  # False Negatives
tp = cm[1][1]  # True Positives

print("\n--- 2. WHAT DO THESE NUMBERS MEAN? ---")
print("=" * 40)
print(f"                 Pred FAIL(0) | Pred PASS(1)")
print(f"Actual FAIL(0) : TN = {tn:<7} | FP = {fp}")
print(f"Actual PASS(1) : FN = {fn:<7} | TP = {tp}")
print("=" * 40)

print("\nPlain English Breakdown:")
print(f"\u2705 We correctly failed {tn} students, and correctly passed {tp} students.")
print(f"🚨 We made {fp} False Positives (passed students who actually failed). False Alarm!")
print(f"🚨 We made {fn} False Negatives (failed students who actually passed). We missed them!")
```
