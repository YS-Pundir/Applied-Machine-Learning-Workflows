```python
# importing the tools\
import pandas as pd 
import numpy as np
import plotly as px

# importing the machine leanring tools
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
# importing the evaluation tools
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
print("Tools imported")
```

```python
from sklearn.datasets import load_breast_cancer

# 1. Load Data & Split
cancer_data = load_breast_cancer(as_frame=True)
X_med = cancer_data.data
y_med = cancer_data.target
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_med, y_med, test_size=0.2, random_state=42)
```

```python
#training model with desicion tree
pipe=make_pipeline(
    MinMaxScaler(),
    DecisionTreeClassifier(max_depth=3,random_state=42)
)
pipe.fit(X_train_m,y_train_m)
prediction=pipe.predict(X_test_m)
print("----- Accuracy Scores -----")
print("The training Score : ",pipe.score(X_test_m,y_test_m)*100,"%")
print("The testing score : ",pipe.score(X_test_m,y_test_m)*100,"%")
print("----- Confusion matrix -----")
matrix=confusion_matrix(prediction,y_test_m)
print(matrix)
tn=matrix[0][0]
fp=matrix[0][1]
fn=matrix[1][0]
tp=matrix[1][1]
print()
print(f"tn : {fn} | fp : {tp}\nfn : {tn}  | tp : {fp}")
print("--------- Classification report ---------")
print(classification_report(prediction,y_test_m))
print("----- ROC-AUC -----")
# Get probabilities for the positive class
probab = pipe.predict_proba(X_test_m)[:, 1]

# Correct order: (y_true, y_score)
roc_score = roc_auc_score(y_test_m, probab)

print(f"ROC AUC Score: {roc_score:.4f}")
```

```python
#training model with desicion tree
pipe=make_pipeline(
    MinMaxScaler(),
    RandomForestClassifier(n_estimators=100,max_depth=3,random_state=42)
)
pipe.fit(X_train_m,y_train_m)
prediction=pipe.predict(X_test_m)
print("----- Accuracy Scores -----")
print("The training Score : ",pipe.score(X_test_m,y_test_m)*100,"%")
print("The testing score : ",pipe.score(X_test_m,y_test_m)*100,"%")
print("----- Confusion matrix -----")
matrix=confusion_matrix(prediction,y_test_m)
print(matrix)
tn=matrix[0][0]
fp=matrix[0][1]
fn=matrix[1][0]
tp=matrix[1][1]
print()
print(f"tn : {fn} | fp : {tp}\nfn : {tn}  | tp : {fp}")
print("--------- Classification report ---------")
print(classification_report(prediction,y_test_m))
print("----- ROC-AUC -----")
# Get probabilities for the positive class
probab = pipe.predict_proba(X_test_m)[:, 1]

# Correct order: (y_true, y_score)
roc_score = roc_auc_score(y_test_m, probab)

print(f"ROC AUC Score: {roc_score:.4f}")
```
