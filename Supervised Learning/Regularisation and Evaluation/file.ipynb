```python
# Step 1 : Importing the tools
import pandas as pd 
import numpy as np 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler,OneHotEncoder,LabelEncoder
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import mean_absolute_error,mean_squared_error,accuracy_score,classification_report,confusion_matrix
from sklearn.pipeline import make_pipeline
from sklearn.datasets import make_regression, fetch_california_housing
```

```python
# 1. Create a tricky dataset with 80 features, but only 5 are actually useful!
X_data, y_data = make_regression(n_samples=100, n_features=80, n_informative=5, noise=20, random_state=42)

# Convert to a Pandas DataFrame so it's easy to read
feature_names = [f"Feature_{i}" for i in range(1, 81)]
X = pd.DataFrame(X_data, columns=feature_names)
y = pd.Series(y_data, name="House_Price")

print("--- OUR TRICKY DATASET ---")
print(f"We have {X.shape[0]} houses and {X.shape[1]} features (clues).")
display(X.head())
```

```python
# 2. Split the Data FIRST
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 3. Scale the Data (Crucial step so Regularization works fairly later)
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Train Standard Linear Regression
model_lr = LinearRegression()
model_lr.fit(X_train_scaled, y_train)

print("--- STANDARD LINEAR REGRESSION SCORES ---")
print(f"Training Score: {model_lr.score(X_train_scaled, y_train):.2f}")
print(f"Testing Score:  {model_lr.score(X_test_scaled, y_test):.2f}")

print("\nVERDICT: CATASTROPHIC OVERFITTING! It scored a perfect 1.00 on Train, but crashed into negative numbers on the completely unseen Test data.")
```

```python
# the features' weight now look like this 
df=pd.DataFrame({"Feature":X.columns,"Learned Values":model_lr.coef_})
df
```

```python
# creating the ridge model 
model_ridge=Ridge(alpha=0.2)
model_ridge.fit(X_train_scaled,y_train)
# as we can see that the ridge has panalized the features to regulate the model
print("The ridge model training score : ",model_ridge.score(X_train_scaled,y_train))
print("The ridge model testign score : ",model_ridge.score(X_test_scaled,y_test))

# the features' weight now look like this 
df=pd.DataFrame({"Feature":X.columns,"Learned Values":model_ridge.coef_})
df
```
