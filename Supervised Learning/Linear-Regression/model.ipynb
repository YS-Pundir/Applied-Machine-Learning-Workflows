```python
import pandas as pd 
import numpy as np
import plotly.express as px

from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

```

```python
df=pd.read_csv("healthcare-dataset-stroke-data.csv")
data=df.copy()
display(data)
```

```python
#Inspecting the data 
# step 1 : Structural info:
print("The Structural Info :  ")
print(data.info())
print("The Arthematic summary : ")
display(data.describe())
print("-"*60)
# Step 2 : Missing data 
print("Inspecting the missing values in data : ")
print(data.isnull().sum())
# step 3 Inspeccting the distribution 
print("_"*60)
fig=px.histogram(
    data,
    x="age",
    title="The Age distribution chart"
)
fig.show()
fig=px.histogram(
    data,
    x="avg_glucose_level",
    title="The Glucose Level distribution chart"
)

fig=px.histogram(
    data,
    x="gender",
    title="Distribution of genders"
    
)
fig.show()
print(data["gender"].value_counts())
fig.show()

#step 4 : Detecting Outliers
fig=px.box(
    data,
    y=["age","avg_glucose_level"],
    title="Outliers Chart"
)
fig.show()
#step 5 : Inspecting the coorelation breakpoint
corr=data.corr(numeric_only=True)
fig=px.imshow(
    corr,
    title="The coorelation chart"
)
fig.show()
```

```python
# droping the problematic features
data=data.drop(["id","bmi"],axis=1)
data=data[data["gender"]!="Other"]

features = [

    "gender", 
    "age", 
    "hypertension", 
    "heart_disease", 
    "ever_married", 
    "work_type", 
    "Residence_type", 
    "avg_glucose_level", 

    "smoking_status", 
  
]

x=data[features]
y=data["stroke"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
# This keeps 'age', 'bmi', etc., and expands the categories into new columns
x_train=pd.get_dummies(x_train,dtype=int,columns=["gender","ever_married","work_type","Residence_type","smoking_status"])
x_test=pd.get_dummies(x_test,dtype=int,columns=["gender","ever_married","work_type","Residence_type","smoking_status"])
```

```python

industry_pipeline = make_pipeline(
    MinMaxScaler(),
    LogisticRegression()
)

```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE ONLY to the training data to prevent leakage
smote = SMOTE(random_state=42)
x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)

# 2. Fit the Pipeline on the RAW training data (It handles the scaling internally!)
industry_pipeline.fit(x_train_smote, y_train_smote)

print("--- PIPELINE EVALUATION ---")
print(f"Training Score: {industry_pipeline.score(x_train_smote, y_train_smote):.2f}")
print(f"Testing Score:  {industry_pipeline.score(x_test, y_test):.2f}")
print("\nAmazing! The pipeline does the exact same math, but with much cleaner and safer code.")
pred=industry_pipeline.predict(x_test)
print(confusion_matrix(y_test,pred))
```
