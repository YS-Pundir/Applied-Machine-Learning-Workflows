```python
# 1. Create a dataset of 200 students (using code from our lecture notes)
rng = np.random.default_rng(seed=42)

# Study hours between 1 and 10
study_hours = rng.uniform(low=1, high=10, size=200)

# Formula: Base score of 40 + (7.5 marks per hour) + some random noise
scores = 40 + 7.5 * study_hours + rng.normal(loc=0, scale=5, size=200)

# Let's put this into a clean Pandas DataFrame so we can view it
df_students = pd.DataFrame({
    'Study_Hours': study_hours,
    'Exam_Score': scores
})

```
