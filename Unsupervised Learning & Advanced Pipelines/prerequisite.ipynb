## Centroid in Machine Learning

A **centroid** is the central point of a cluster of data points in a dataset. It represents the **average position** of all the points in that cluster in a multi-dimensional space.

In machine learning, centroids are mainly used in clustering algorithms such as K-Means to group similar data points together. The centroid is calculated by taking the **mean of all feature values** of the data points in a cluster.

### Mathematical Formula

For \( n \) data points:

\[
C = \frac{1}{n} \sum_{i=1}^{n} x_i
\]

Where:
- \( C \) = centroid  
- \( n \) = number of data points  
- \( x_i \) = each data point  

### Key Points
- A centroid represents the **center of a cluster**.
- It is calculated as the **mean of all data points** in that cluster.
- Centroids are updated iteratively in clustering algorithms.
- They help in grouping **similar data points together**.

### Example
If a cluster contains points (2,4), (4,6), and (6,8), the centroid is:

\[
(4,6)
\]

This point represents the **average location of the cluster**.

## Euclidean Distance

**Euclidean Distance** is a measure used to calculate the **straight-line distance between two points** in a geometric space. It is one of the most common distance metrics used in machine learning, data mining, and pattern recognition.

In machine learning, Euclidean distance is often used to measure the **similarity between data points**, especially in algorithms like clustering and classification.

### Mathematical Formula

For two points \( P(x_1, y_1) \) and \( Q(x_2, y_2) \):

\[
d(P,Q) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
\]

Where:
- \( d(P,Q) \) = Euclidean distance between points \(P\) and \(Q\)
- \( (x_1, y_1) \) = coordinates of the first point
- \( (x_2, y_2) \) = coordinates of the second point

### General Formula (n-dimensional)

For two points \( X(x_1, x_2, ..., x_n) \) and \( Y(y_1, y_2, ..., y_n) \):

\[
d(X,Y) = \sqrt{\sum_{i=1}^{n}(x_i - y_i)^2}
\]

### Key Points
- Measures the **shortest straight-line distance** between two points.
- Commonly used in **clustering and nearest neighbor algorithms**.
- Works in **2D, 3D, or multi-dimensional space**.

### Example

For points \( A(1,2) \) and \( B(4,6) \):

\[
d = \sqrt{(4-1)^2 + (6-2)^2}
\]

\[
d = \sqrt{9 + 16}
\]

\[
d = \sqrt{25} = 5
\]

So, the Euclidean distance between the two points is **5**.

## Variance and Spread

### Spread of Data
**Spread** refers to how much the values in a dataset differ from each other and from the mean.  
It tells us **how widely the data are distributed**.

If the data values are **close to the mean**, the spread is small.  
If the data values are **far from the mean**, the spread is large.

Common measures of spread include:
- Range
- Variance
- Standard Deviation
- Interquartile Range

---

## Variance

**Variance** is a numerical measure of the **dispersion of data values from their mean**.  
It is calculated as the **average of the squares of deviations from the mean**.

### Mathematical Formula (NCERT Style)

For observations \(x_1, x_2, x_3, ..., x_n\) with mean \(\bar{x}\):

\[
\sigma^2 = \frac{1}{n}\sum_{i=1}^{n}(x_i - \bar{x})^2
\]

Where:
- \( \sigma^2 \) = variance  
- \( x_i \) = each observation  
- \( \bar{x} \) = mean of the observations  
- \( n \) = number of observations  

---

### Step-wise Method (as used in NCERT)

1. Find the **mean**

\[
\bar{x} = \frac{\sum x_i}{n}
\]

2. Find **deviation from mean**

\[
x_i - \bar{x}
\]

3. Square the deviations

\[
(x_i - \bar{x})^2
\]

4. Find the **average of squared deviations**

\[
\sigma^2 = \frac{\sum (x_i - \bar{x})^2}{n}
\]

---

### Example

Dataset: 2, 4, 6, 8

Mean:

\[
\bar{x} = \frac{2 + 4 + 6 + 8}{4} = 5
\]

Squared deviations:

\[
(2-5)^2 = 9
\]

\[
(4-5)^2 = 1
\]

\[
(6-5)^2 = 1
\]

\[
(8-5)^2 = 9
\]

Variance:

\[
\sigma^2 = \frac{9 + 1 + 1 + 9}{4} = 5
\]

Thus, the **variance of the dataset is 5**, which represents the spread of the data around the mean.
